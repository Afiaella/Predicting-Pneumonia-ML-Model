```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('causes-of-death-in-children.csv')
```

```python
df.head()
```

```python
to_plot = ["Deaths - Malaria - Sex: Both - Age: Under 5 (Number)", 
           "Deaths - HIV/AIDS - Sex: Both - Age: Under 5 (Number)",
           "Deaths - Lower respiratory infections - Sex: Both - Age: Under 5 (Number)", "Deaths - Whooping cough - Sex: Both - Age: Under 5 (Number)",
          "Deaths - Nutritional deficiencies - Sex: Both - Age: Under 5 (Number)", "Deaths - Drowning - Sex: Both - Age: Under 5 (Number)",
          "Deaths - Syphilis - Sex: Both - Age: Under 5 (Number)"]

labels = ["Malaria", "HIV/AIDS", "Pneumonia", "Whooping Cough", "Nutritional deficiencies", "Drowning", "Syphilis"]
```

```python
fig, ax = plt.subplots(figsize=(8,6))

for col, label in zip(to_plot,labels):
    sns.lineplot(x='Year', y = col, data=df.groupby(by='Year').sum().reset_index(), label=label)
    
plt.savefig('my_figure.png', )
```
